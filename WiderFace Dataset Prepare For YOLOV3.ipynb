```python
from PIL import Image
import numpy as np
import os
from os import listdir
from os.path import isfile, join

def parseImFilename(imFilename, imPath):
    
    #Open the Image given by imFilename and Path @ imPath
    
    im = Image.open(os.path.join(imPath, imFilename))  
    
    #Get the Height and width for the image.
    width, height = im.size
    
    #new_File = open("D:/WiderFace/output/" + os.path.splitext(imFilename)[0] +".txt","w+")
     
    #Open the WiderFace split
    with open("D:/WiderFace/wider_face_split/wider_face_split/wider_face_train_bbx_gt.txt") as myFile:
        for num, line in enumerate(myFile, 1):
                if imFilename in line:
                    
                    myNumber = myFile.readline()
                    #print ('Number_of_classes', myNumber)
                    np.set_printoptions(precision=6)
                    for i in range (int(myNumber)):
                        temp = np.array(myFile.readline().split()).astype(np.float)
                        dx = round((temp[0]+(temp[2]/2))/width,6)
                        dy = round((temp[1]+(temp[3]/2))/height,6)
                        dw = round(temp[2]/width,6)
                        dh = round(temp[3]/height,6)
                        new_File_line = "0 %s %s %s %s\n" % (dx,dy,dw,dh)
                        print(new_File_line)
                        #new_File.write(new_File_line)
                    
                    #new_File.close()
    return imFilename, str(width), str(height)

def listDiectory(imPath):
    
    onlyfiles = [f for f in listdir(imPath) if isfile(join(imPath, f))]
    for i in range (len(onlyfiles)):
        print(onlyfiles[i])
        parseImFilename(onlyfiles[i], imPath)
        
def printDirectory(imPath):
    onlyfiles = [f for f in listdir(imPath) if isfile(join(imPath, f))]
    for i in range (len(onlyfiles)):
        print("/home/liuyan/Desktop/darknet/data/sample_train/output/" + onlyfiles[i])
        


    
```
